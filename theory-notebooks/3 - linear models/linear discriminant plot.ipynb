```python
from matplotlib import pyplot as plt
import numpy as np

x1 = np.linspace(-20,20,1000)
x2 = np.linspace(-20,20,1000)
x1,x2 = np.meshgrid(x1,x2)

# just some randomly initialized weights
w0 = 3
w = [2,2]

# linear discriminator function
y = np.zeros((len(x1), len(x2)))
for i in range(len(x1)):
    for j in range(len(x2)):
        y[j,i] = w0 + w[0]*x1[j,i] + w[1]*x2[j,i]
plt.contour(x1,x2,y,levels=[w0+w[0]**2+w[1]**2],colors="k")
plt.contour(x1,x2,y,levels=[-w0/w[0]**2+w[1]**2],colors="r")

plt.text(*w, s='w', ha='right')
plt.plot([0,w[0]],[0,w[1]],"-o")
plt.plot([0,w[1]],[0,-w[0]],"-o")
plt.axis("equal")
plt.axis([-15,15,-15,15])
plt.grid()

# let's classify a point
sample = [-1,-1] # classified negatively because it lies on the opposite part wrt w vector
y = np.dot(sample,w) + w0
plt.plot(*sample,"og")
print(y) 

plt.show()
```
